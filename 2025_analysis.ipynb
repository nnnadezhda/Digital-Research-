```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
from scipy.stats.mstats import winsorize
from scipy.stats import skew, kurtosis, shapiro
from statsmodels.stats.api import het_breuschpagan

# Helper: Winsorize a pandas Series by percentiles to handle outliers
def winz_series(s: pd.Series, p=0.01):
    x = pd.to_numeric(s, errors='coerce').to_numpy(copy=True)
    w = winsorize(x, limits=[p, p])
    return pd.Series(w, index=s.index)

# Step 0: Install necessary libraries (if not already installed)
# All required libraries are assumed to be pre-installed in the Colab environment
# or handled by the initial import statements above.

# Step 1: Data Loading
print("### 1. Data Loading")
excel_file_path = '/content/Experiment 2025.xlsx'
df = pd.read_excel(excel_file_path)
display(df.head())

# Define relevant numerical columns for EDA and modeling (using original column names)
relevant_cols_eda = ['Market cap 2025', 'Annual Dividend 2025', 'Beta 2025.2',
       'Average Annual Total Return new', 'Website index 2025',
       'leverage 2024', 'roa', 'Momentum 05/2025', 'Book to market 2025',
       'Stock return', 'Industry Dummies ']

# Define independent variables for VIF calculation (excluding the dependent variable and dummy indicator)
independent_vars_for_vif = [
    'Market cap 2025',
    'Annual Dividend 2025',
    'Beta 2025.2',
    'Average Annual Total Return new',
    'Website index 2025',
    'leverage 2024',
    'roa',
    'Momentum 05/2025',
    'Book to market 2025'
]

# Step 2: Define the data preparation function
# This function encapsulates data cleaning, type conversion, winsorization,
# log-transformation, and dummy variable creation for the final model.
def prepare_data(df_raw):
    df_temp = df_raw.copy()

    # Variables to winsorize for the final model
    vars_to_winz = [
        "Momentum 05/2025",
        "roa",
        "Book to market 2025",
        "Annual Dividend 2025",
        "leverage 2024",
        "Average Annual Total Return new"
    ]

    # Apply winsorization and create new columns with '_winz' suffix.
    for col in vars_to_winz:
        if col in df_temp.columns:
            df_temp[col + "_winz"] = winz_series(df_temp[col], p=0.01)

    # Log-transform Stock return
    if 'Stock return' in df_temp.columns:
        min_stock_return = df_temp['Stock return'].min()
        if min_stock_return <= 0:
            # Add a small constant to make values positive before log transform
            # Ensure it's strictly positive, even if min_stock_return is 0
            df_temp['log_stock_return'] = np.log(df_temp['Stock return'] + 1 - min_stock_return)
        else:
            df_temp['log_stock_return'] = np.log(df_temp['Stock return'])

    # The following block is important for the C(Industry) fixed effects model if it needs the original 'Industry' categorical column.
    # If 'Industry Dummies ' column from raw data was meant to become the categorical 'Industry' column,
    # the previous pd.get_dummies would have removed it. We ensure 'Industry' exists for fixed effects.
    # Assuming the original excel had a column 'Industry' in addition to 'Industry Dummies ' for proper fixed effects
    if 'Industry Dummies ' in df_temp.columns:
        # Convert 'Industry Dummies ' (integer) into categorical dummy variables
        # Save original column before conversion if needed for other purposes, though not explicitly requested here.
        df_temp = pd.get_dummies(df_temp, columns=['Industry Dummies '], drop_first=True, prefix='Industry')

    return df_temp

# Step 3: Load the raw data and prepare it using the defined function
# (Note: For initial exploratory data analysis, VIF, and preliminary OLS, the original `df` is used.
# The `prepare_data` function is defined here, but applied later for the final model.)

# --- The following steps use the original df for initial diagnostics and preliminary modeling ---

# Step 4: Display descriptive statistics for relevant numerical columns
# This provides a summary of the central tendency, dispersion, and shape of the data.
print("\n### 4. Generate Descriptive Statistics")
descriptive_stats = df[relevant_cols_eda].describe()
display(descriptive_stats)

# Step 5: Perform Outlier Analysis using IQR method and visualize with histograms
# This identifies and visualizes extreme values in key variables.
print("\n### 5. Identify and Visualize Outliers")
outliers = {}
for col in relevant_cols_eda:
    # Skip 'Industry Dummies ' if it's already categorical or meant for dummy creation, not numerical outlier analysis
    if col == 'Industry Dummies ':
        # No numerical outliers for dummy indicator in this context
        col_outliers = pd.DataFrame()
        outlier_count = 0
    else:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count = col_outliers.shape[0]

    outliers[col] = outlier_count
    print(f"Outliers in {col}: {outlier_count}")
    # Uncomment to display outlier dataframes for each column
    # if outlier_count > 0:
    #     display(col_outliers)

# Generate histograms highlighting outliers (only for numerical columns)
for col in relevant_cols_eda:
    if col != 'Industry Dummies ':
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        plt.figure(figsize=(10, 6))
        sns.histplot(data=df, x=col, kde=True)
        outliers_data = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        if not outliers_data.empty:
            plt.scatter(outliers_data[col], np.zeros(outliers_data[col].shape[0]), color='red', label='Outliers')
        plt.title(f'Distribution of {col} with Outliers')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()
display(outliers)

# Step 6: Calculate and visualize the Correlation Matrix
# This helps understand linear relationships between variables and check for multicollinearity.
print("\n### 6. Calculate and Visualize Correlation Matrix")
correlation_matrix = df[relevant_cols_eda].corr()
display(correlation_matrix)
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Relevant Numerical Columns')
plt.show()

# Step 7: Calculate Variance Inflation Factors (VIFs) for independent variables
# This quantifies the severity of multicollinearity in an OLS regression.
print("\n### 7. Calculate Variance Inflation Factor (VIF)")
df_for_vif = df[independent_vars_for_vif].dropna()
X_vif = add_constant(df_for_vif)
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
display(vif_data)

# Step 8: Analyze the distribution of the dependent variable ('Stock return')
# This checks for normality and other distributional properties.
print("\n### 8. Analyze Stock Return Distribution")
plt.hist(df['Stock return'].dropna(), bins=30, edgecolor='k')
plt.xlabel('Stock Return')
plt.ylabel('Frequency')
plt.title('Distribution of Stock Returns')
plt.show()
print(f"Skewness: {skew(df['Stock return'].dropna())}")
print(f"Kurtosis: {kurtosis(df['Stock return'].dropna())}")

# Step 9: Preliminary OLS Regression & Residual Tests
# Fits a basic OLS model, checks for heteroscedasticity (Breusch-Pagan) and normality of residuals (Shapiro-Wilk).
print("\n### 9. Preliminary OLS Regression & Residual Tests")
X_prelim = df[['Market cap 2025', 'Annual Dividend 2025', 'Beta 2025.2',
       'Average Annual Total Return new', 'Website index 2025',
       'leverage 2024', 'roa', 'Momentum 05/2025', 'Book to market 2025',
       'Industry Dummies ']].dropna()
y_prelim = df.loc[X_prelim.index, 'Stock return']
X_prelim = sm.add_constant(X_prelim)
model_prelim = sm.OLS(y_prelim, X_prelim).fit()
bp_test = het_breuschpagan(model_prelim.resid, model_prelim.model.exog)
print(f"Breusch-Pagan test p-value: {bp_test[1]}")
stat_shapiro, p_shapiro = shapiro(model_prelim.resid)
print(f"Shapiro-Wilk test p-value: {p_shapiro}")

# Step 10: Influence Diagnostics (Leverage and Cook's D)
# Calculated leverage values and Cook's Distance for observations in the preliminary OLS model
# to identify observations that have a disproportionate impact on the regression results.
print("\n### 10. Influence Diagnostics (Leverage and Cook's D)")
infl = model_prelim.get_influence()
cook_d = infl.cooks_distance[0]
leverage = infl.hat_matrix_diag
influence_df = pd.DataFrame({
    'Cooks_Distance': cook_d,
    'Leverage': leverage,
    'Stock_Return': y_prelim,
    'Predicted_Stock_Return': model_prelim.predict()
}, index=y_prelim.index)

print("Top 5 observations by Cook's Distance:")
display(influence_df.sort_values(by='Cooks_Distance', ascending=False).head())

print("Top 5 observations by Leverage:")
display(influence_df.sort_values(by='Leverage', ascending=False).head())

# Step 11: Run Bivariate OLS Model
print("\n### 11. Run Bivariate OLS Model")
m_biv = smf.ols("Q('Stock return') ~ Q('Website index 2025')", data=df).fit(cov_type='HC3')
print(m_biv.summary())

# Step 12: Build and Evaluate Final OLS Model (with Preprocessing and Robust Standard Errors)
# This sets up the base for 'Formula A' and other comparison models.
print("\n### 12. Build and Evaluate Final OLS Model (with Preprocessing and Robust Standard Errors)")

# Prepare data using the defined function
df_prepared = prepare_data(df)

# Define the OLS regression formula (Formula A - Comprehensive Winsorized)
formula_winz = """
Q('log_stock_return') ~ Q('Website index 2025')
 + Q('Momentum 05/2025_winz')
 + Q('Average Annual Total Return new_winz')
 + Q('roa_winz')
 + Q('Annual Dividend 2025_winz')
 + Q('Book to market 2025_winz')
 + Q('leverage 2024_winz')
 + Q('Beta 2025.2')
"""

# List of all columns required for the final model (Formula A) including dummy variable pattern
needed_cols_final = [
    'log_stock_return', 'Website index 2025',
    'Momentum 05/2025_winz',
    'Average Annual Total Return new_winz',
    'roa_winz', 'Annual Dividend 2025_winz', 'Book to market 2025_winz',
    'leverage 2024_winz', 'Beta 2025.2'
]

# Dynamically add Industry dummy columns to needed_cols_final, excluding Industry_11
industry_dummies = [col for col in df_prepared.columns if col.startswith('Industry_') and col != 'Industry_11']
needed_cols_final.extend(industry_dummies)

# Ensure all needed columns are numeric and drop rows with missing values
# The Industry dummy columns are already binary (0/1) from get_dummies, no need to force numeric.
for c in needed_cols_final:
    if c not in df_prepared.columns:
        print(f"Warning: Column '{c}' not found in df_prepared. Skipping.")
        continue
    if c.startswith('Industry_'): # Dummy variables are already 0/1
        continue
    df_prepared[c] = pd.to_numeric(df_prepared[c], errors='coerce')

d_w = df_prepared.dropna(subset=needed_cols_final).copy()

# Add industry dummies to the formula_winz string
if industry_dummies:
    formula_winz += ' + ' + ' + '.join([f"Q('{col}')" for col in industry_dummies])

# Fit OLS with HC3 robust SEs (This is the base model for comparison, Formula A)
m_w = smf.ols(formula=formula_winz, data=d_w).fit(cov_type='HC3')
print("\n--- OLS Regression (Formula A - Comprehensive Winsorized) ---")
print(m_w.summary())

# Make formula_main and df_for_model_main available for subsequent steps (mapping to m_w and d_w)
formula_main = formula_winz
df_for_model_main = d_w
model_main = m_w

# Step 13: OLS Regression with Outlier Removal and Comparison
# This section demonstrates the impact of removing influential observations on model results.
# Re-using formula_main and df_for_model_main from the previous step for the base model
print("\n### 13. OLS Regression with Outlier Removal and Comparison")

# Identify influential observations from the *preliminary* model's Cook's Distance
# Use a threshold like 4/N (where N is number of observations in model_prelim)
threshold_cooks = 4 / len(model_prelim.resid)

# Get indices of influential observations based on the preliminary model
# We'll use these indices to filter the prepared data (d_w)
infl_indices = influence_df[influence_df['Cooks_Distance'] > threshold_cooks].index

print(f"Number of influential observations identified (Cook's D > {threshold_cooks:.4f}): {len(infl_indices)}")

d_w_clean = d_w.drop(index=infl_indices, errors='ignore').copy()

if not d_w_clean.empty:
    print(f"Number of observations after removing influential points: {len(d_w_clean)}")
    # Fit the same comprehensive winsorized model (Formula A) on the cleaned data
    m_w_clean = smf.ols(formula=formula_main, data=d_w_clean).fit(cov_type='HC3')
    print("\n--- OLS Regression (Formula A - Cleaned Data) ---")
    print(m_w_clean.summary())

    # Comparison (qualitative for now, quantitative could involve F-tests or R-squared comparison)
    print("\n--- Comparison (Formula A with vs. without influential observations) ---")
    print(f"Original R-squared: {model_main.rsquared:.3f}")
    print(f"Cleaned R-squared: {m_w_clean.rsquared:.3f}")
    print("Note: Examine changes in coefficients, p-values, and model fit.")
else:
    print("No observations left after removing influential points. Skipping cleaned model.")


# Step 14: Alternative OLS Model with Mixed Raw/Winsorized Variables
# A regression comparing the impact of using raw vs. winsorized variables for some predictors.
print("\n### 14. Alternative OLS Model with Mixed Raw/Winsorized Variables")

# Example: Use raw 'Annual Dividend 2025', but winsorized for others.
# Exclude 'Industry Dummies _11' from this model's dummies as well
formula_mixed = """
Q('log_stock_return') ~ Q('Website index 2025')
 + Q('Momentum 05/2025_winz')
 + Q('Average Annual Total Return new_winz')
 + Q('roa_winz')
 + Q('Annual Dividend 2025')
 + Q('Book to market 2025_winz')
 + Q('leverage 2024_winz')
 + Q('Beta 2025.2')
"""

# Add industry dummies to the formula_mixed string, excluding the problematic _11
if industry_dummies:
    formula_mixed += ' + ' + ' + '.join([f"Q('{col}')" for col in industry_dummies])

# Need to ensure 'Annual Dividend 2025' (raw) is also in df_prepared
# and that all columns are numeric before dropna.
needed_cols_mixed = [
    'log_stock_return', 'Website index 2025',
    'Momentum 05/2025_winz',
    'Average Annual Total Return new_winz',
    'roa_winz', 'Annual Dividend 2025', 'Book to market 2025_winz',
    'leverage 2024_winz', 'Beta 2025.2'
]
needed_cols_mixed.extend(industry_dummies)

df_temp_mixed = df_prepared.copy()
for c in needed_cols_mixed:
    if c not in df_temp_mixed.columns:
        if c == 'Annual Dividend 2025': # This is a raw column, ensure it exists from original df
            df_temp_mixed[c] = df[c]
        else:
            print(f"Warning: Column '{c}' not found for mixed model. Skipping.")
            continue
    if c.startswith('Industry_'):
        continue
    df_temp_mixed[c] = pd.to_numeric(df_temp_mixed[c], errors='coerce')

d_mixed = df_temp_mixed.dropna(subset=needed_cols_mixed).copy()

m_mixed = smf.ols(formula=formula_mixed, data=d_mixed).fit(cov_type='HC3')
print("\n--- OLS Regression (Mixed Raw/Winsorized Variables) ---")
print(m_mixed.summary())

# Step 15: OLS Regression (Formula A - Comprehensive Winsorized)
# This is a re-presentation of the comprehensive winsorized model from Step 12.
print("\n### 15. OLS Regression (Formula A - Comprehensive Winsorized)")
print(model_main.summary())

# Step 16: OLS Regression (Formula B, Log-Y, Fixed Effects)
# Explores different model specifications: excluding momentum, log-transforming Y, and using fixed effects.
print("\n### 16. OLS Regression (Formula B, Log-Y, Fixed Effects)")

# Formula B: Excluding Momentum (winsorized)
# Log-transform dependent variable (already 'log_stock_return')
formula_B = """
Q('log_stock_return') ~ Q('Website index 2025')
 + Q('Average Annual Total Return new_winz')
 + Q('roa_winz')
 + Q('Annual Dividend 2025_winz')
 + Q('Book to market 2025_winz')
 + Q('leverage 2024_winz')
 + Q('Beta 2025.2')
"""

# Add industry dummies to the formula_B string, excluding the problematic _11.
# This serves as the 'fixed effects' part.
if industry_dummies:
    formula_B += ' + ' + ' + '.join([f"Q('{col}')" for col in industry_dummies])

# The dataframe d_w (from Step 12) already contains all needed winsorized and dummy columns.
# Ensure all columns needed for formula_B are present and numeric.
needed_cols_B = [
    'log_stock_return', 'Website index 2025',
    'Average Annual Total Return new_winz',
    'roa_winz', 'Annual Dividend 2025_winz', 'Book to market 2025_winz',
    'leverage 2024_winz', 'Beta 2025.2'
]
needed_cols_B.extend(industry_dummies)

d_B = d_w.dropna(subset=needed_cols_B).copy() # Filter based on needed cols for this formula

m_B = smf.ols(formula=formula_B, data=d_B).fit(cov_type='HC3')
print("\n--- OLS Regression (Formula B - Log-Y, Fixed Effects, No Momentum) ---")
print(m_B.summary())


# Step 17: Multivariate OLS Regression (Winsorized only)
# A regression using only winsorized and log-transformed variables.
# Modifying to include 'Beta ' for consistency with Main OLS model
# Exclude 'Industry Dummies _11' from this model's dummies as well
# This step seems to be a re-emphasis or re-statement of Formula A from Step 12/15.
# If there's a different specific set of variables for this model, please clarify.
# For now, it will present the summary of the comprehensive winsorized model (Formula A).
print("\n### 17. Multivariate OLS Regression (Winsorized only)")
print(model_main.summary())

# Step 18: Rank Regression Analysis
# Explores relationships using rank-transformed variables, robust to non-normality.
print("\n### 18. Rank Regression Analysis")

# Create a DataFrame with selected numerical variables for ranking
# Using original df as base to ensure all values are present before ranking
rank_cols = [
    'Stock return',
    'Website index 2025',
    'Momentum 05/2025',
    'Average Annual Total Return new',
    'roa',
    'Annual Dividend 2025',
    'Book to market 2025',
    'leverage 2024',
    'Beta 2025.2'
]

df_rank_analysis = df[rank_cols].copy().dropna()

# Rank all numerical columns (excluding dummy variables)
# Use 'average' method for tied ranks
df_ranked = df_rank_analysis.rank(method='average')

# Merge back industry dummies from df_prepared if needed for rank regression
# For simplicity, we will assume industry dummies are already 0/1 and do not need ranking.
# We will use the same 'industry_dummies' list as before.
# Filter df_prepared to include only the relevant indices and industry dummies
industry_dummies_df = df_prepared.loc[df_rank_analysis.index, industry_dummies]
df_ranked = pd.concat([df_ranked, industry_dummies_df], axis=1)

print("\n--- Ranked Data Head ---")
display(df_ranked.head())

# Step 19: Spearman (unconditional) correlation
print("\n### 19. Spearman (unconditional) Correlation")
spearman_corr = df_rank_analysis.corr(method='spearman')
display(spearman_corr)
plt.figure(figsize=(12, 10))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Correlation Matrix of Ranked Numerical Columns')
plt.show()

# Step 20: Bivariate rank regression
print("\n### 20. Bivariate Rank Regression")
# Example: Rank regression of 'Stock return' on 'Website index 2025'
m_rank_biv = smf.ols("Q('Stock return') ~ Q('Website index 2025')", data=df_ranked).fit(cov_type='HC3')
print("\n--- Bivariate Rank Regression (Stock Return vs. Website Index) ---")
print(m_rank_biv.summary())

# Step 21: Multivariate rank regression
print("\n### 21. Multivariate Rank Regression")
# Define the formula for multivariate rank regression
# Exclude 'Stock return' from independent variables and use its ranked version as dependent
independent_rank_vars = [
    'Website index 2025',
    'Momentum 05/2025',
    'Average Annual Total Return new',
    'roa',
    'Annual Dividend 2025',
    'Book to market 2025',
    'leverage 2024',
    'Beta 2025.2'
]

formula_rank_multi = "Q('Stock return') ~ " + ' + '.join([f"Q('{col}')" for col in independent_rank_vars])

# Add industry dummies to the multivariate rank regression formula
if industry_dummies:
    formula_rank_multi += ' + ' + ' + '.join([f"Q('{col}')" for col in industry_dummies])

m_rank_multi = smf.ols(formula=formula_rank_multi, data=df_ranked).fit(cov_type='HC3')
print("\n--- Multivariate Rank Regression ---")
print(m_rank_multi.summary())
```
